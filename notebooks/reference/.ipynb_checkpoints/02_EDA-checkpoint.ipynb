# Find the potential competitors for our client in each theme

```python
merge_sales_data.head()
```

```python
merge_sales_data.theme_id.nunique()
```

```python
# Calculate total sales for each vendor within each theme
total_sales = merge_sales_data.groupby(['vendor', 'theme_name'])['sales_lbs_value'].sum().reset_index()

# Compute total sales for 'Manufacturer A' within each theme
manufacturer_A_sales = total_sales[total_sales['vendor'] == 'A']

# Join total sales and manufacturer_A_sales on 'Claim Name'
merged_sales = total_sales.merge(manufacturer_A_sales, on='theme_name', suffixes=('', '_A'))

# Determine competitors: Vendors whose sales are within 30% of 'Manufacturer A'
merged_sales['competitor'] = (merged_sales['sales_lbs_value'] >= 0.7 * merged_sales['sales_lbs_value_A']) & \
                             (merged_sales['sales_lbs_value'] <= 1.3 * merged_sales['sales_lbs_value_A'])

# Filter rows where 'competitor' is True and 'Vendor' is not 'Manufacturer A'
competitors = merged_sales[(merged_sales['competitor']) & (merged_sales['vendor'] != 'A')]

# Print competitors
competitors_by_theme=[]
for theme, group in competitors.groupby('theme_name'):
    competitors_by_theme.append([theme, group['vendor'].tolist()])
    
competitors_df = pd.DataFrame(competitors_by_theme, columns=['theme_name', 'competitors'])
competitors_df
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
def seasonality_plot(theme):

    #theme='blueberry'
    merge_sales_data_theme=merge_sales_data[merge_sales_data.theme_name==theme]
    merge_sales_data_theme = merge_sales_data_theme.groupby(['date', 'theme_name']).agg({'sales_lbs_value': 'sum'}).reset_index()
    merge_sales_data_theme.groupby(['date', 'theme_name']).size().count()

    # Convert the 'date' column to datetime format
    merge_sales_data_theme['date'] = pd.to_datetime(merge_sales_data_theme['date'])

    # Set the 'date' column as the DataFrame's index
    merge_sales_data_theme.set_index('date', inplace=True)

    # Perform seasonal decomposition
    decomposition = seasonal_decompose(merge_sales_data_theme['sales_lbs_value'], model='additive')

    # Plot the trend, seasonal, and residual components
    plt.figure(figsize=(8, 6))

    plt.subplot(311)
    plt.plot(decomposition.trend)
    plt.title('Trend: '+theme)
    plt.xlabel('Date')
    plt.ylabel('Sales (lbs)')

    plt.subplot(312)
    plt.plot(decomposition.seasonal)
    plt.title('Seasonal: '+theme)
    plt.xlabel('Date')
    plt.ylabel('Sales (lbs)')

    plt.tight_layout()
    plt.show()
```

```python
seasonality_plot('low carb')
seasonality_plot('no additives/preservatives')
```

```python
def latency_plot(theme):
    # Convert the date columns to datetime format
    merge_sales_data['date'] = pd.to_datetime(merge_sales_data['date'])
    merge_google_search_data['date'] = pd.to_datetime(merge_google_search_data['date'])
    merge_social_media_data['date'] = pd.to_datetime(merge_social_media_data['date'])

    # Merge the dataframes based on the common date
    merged_data = pd.merge(merge_sales_data, merge_google_search_data, on=['date', 'theme_name'])
    merged_data = pd.merge(merged_data, merge_social_media_data, on=['date', 'theme_name'])

    # Filter data for the given theme
    merged_data_theme = merged_data[merged_data['theme_name'] == theme]

    # Set the date column as the index
    merged_data_theme.set_index('date', inplace=True)

    # Aggregate the data to monthly intervals
    monthly_data = merged_data_theme.resample('M').sum()

    # Set up the figure and axes
    fig, ax = plt.subplots(figsize=(10, 4))

    # Plot Sales lbs Value
    ax.set_xlabel('Date')
    ax.set_ylabel('Sales lbs Value')
    ax.plot(monthly_data.index, monthly_data['sales_lbs_value'], color='tab:green', label='Sales lbs Value')

    # Plot Google Search Volume
    ax2 = ax.twinx()
    ax2.set_ylabel('Google Search Volume')
    ax2.plot(monthly_data.index, monthly_data['google_searchVolume'], color='tab:blue', label='Google Search Volume')

    # Plot Social Media Posts
    ax3 = ax.twinx()
    ax3.spines['right'].set_position(('outward', 80))
    ax3.set_ylabel('Social Media Posts')
    ax3.plot(monthly_data.index, monthly_data['total_post'], color='tab:red', label='Social Media Posts')

    # Adjust spacing between subplots
    fig.tight_layout(pad=3.0)

    # Combine all the legends
    lines = ax.get_lines() + ax2.get_lines() + ax3.get_lines()
    labels = [line.get_label() for line in lines]
    ax.legend(lines, labels)

    # Rotate x-axis labels
    plt.xticks(rotation=45)

    # Show the plot
    plt.title('Monthly Trend for Sales lbs Value, Google Search Volume, and Social Media Posts: ' + theme)
    plt.show()
```

```python
latency_plot('low carb')
latency_plot('no additives/preservatives')
latency_plot('salmon')
latency_plot('blueberry')
latency_plot('chicken')
latency_plot('halal')
```

```python
def correlation(theme):
    print('Theme: ', theme)
    # Merge the three dataframes on the common 'date' and 'theme_name' columns
    merged_data = pd.merge(merge_social_media_data, merge_google_search_data, on=['date', 'theme_name'])
    merged_data = pd.merge(merged_data, merge_sales_data, on=['date', 'theme_name'])

    # Filter the data for the desired theme
    merged_data_theme = merged_data[merged_data['theme_name'] == theme]

    # Convert the date columns to datetime format
    merged_data_theme['date'] = pd.to_datetime(merged_data_theme['date'])

    # Set the date column as the index
    merged_data_theme.set_index('date', inplace=True)

    # Aggregate the data to monthly intervals
    merged_data_monthly = merged_data_theme.resample('M').sum()

    # Calculate the cross-correlation between social media posts and search volume
    cross_corr_search = np.correlate(merged_data_monthly['total_post'], merged_data_monthly['google_searchVolume'], mode='full')

    # Calculate the cross-correlation between search volume and sales
    cross_corr_sales = np.correlate(merged_data_monthly['google_searchVolume'], merged_data_monthly['sales_lbs_value'], mode='full')

    # Calculate the time lags for the cross-correlation
    time_lags = np.arange(-len(merged_data_monthly) + 1, len(merged_data_monthly))

    # Plot the cross-correlation functions
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 4))
    ax1.plot(time_lags, cross_corr_search)
    ax1.set_xlabel('Time Lag')
    ax1.set_ylabel('Cross-Correlation')
    ax1.set_title('Cross-Correlation: Social Media Posts vs. Search Volume')

    ax2.plot(time_lags, cross_corr_sales)
    ax2.set_xlabel('Time Lag')
    ax2.set_ylabel('Cross-Correlation')
    ax2.set_title('Cross-Correlation: Search Volume vs. Sales')

    # Find the time lags with the maximum cross-correlation
    max_corr_search_idx = np.argmax(cross_corr_search)
    time_delay_search = time_lags[max_corr_search_idx]

    max_corr_sales_idx = np.argmax(cross_corr_sales)
    time_delay_sales = time_lags[max_corr_sales_idx]

    print(f"Time delay (in months) - Social Media Posts vs. Search Volume: {time_delay_search}")
    print(f"Time delay (in months) - Search Volume vs. Sales: {time_delay_sales}")

    # Show the plots
    plt.tight_layout()
    plt.show()
```

```python
correlation('low carb')
correlation('no additives/preservatives')
correlation('salmon')
correlation('blueberry')
correlation('chicken')
correlation('halal')
```
